```python
# 2d
import pandas as pd
data = {
    "name": ['nikhil', 'raj', 'devendra', 'hello'],
    'age': [10,20,30,40]
}
df = pd.Series(data) # 1d
print(df)
# 2d
df2 = pd.DataFrame(data)
df2
```

```python
data = [
    ['nikhil', 10],
    ['raj', 20],
    ['devendra', 30],
    ['hello', 40]
]
df = pd.DataFrame(data, columns=['name', 'age'])
print(df)
```

```python
#add indexing name
df = pd.DataFrame(data, index=['a', 'b', 'c', 'd'], columns=['name', 'age'])
print(df)
#
```
