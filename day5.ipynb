```python
def calculator(*nums) -> int:
  operator = nums[-1] # Get the last element as the operator
  values = nums[:-1]  # Get all elements except the last as values

  if operator == '+':
    result = 0
    for num in values:
      result += num
    print(result)
    return result
  elif operator == '-':
    if not values:
      print("Error: No numbers provided for subtraction.")
      return None
    result = values[0]
    for num in values[1:]:
      result -= num
    print(result)
    return result
  elif operator == '*':
    result = 1
    for num in values:
      result *= num
    print(result)
    return result
  else:
    print(f"Error: Unsupported operator '{operator}'")
    return None
```

```python
calculator(10,20, 30, 40, '+')
calculator(10,20, 30, 40, '-')
calculator(10,20, 30, 40, '*')
calculator(10,20, 30, 40, '/')
```

```python
def calculator(*nums) -> int:
  operator = nums[-1] # Get the last element as the operator
  values = nums[:-1]  # Get all elements except the last as values

  if not values:
    print("Error: No numbers provided for calculation.")
    return None

  # Construct the expression string. Example: '10+20+30+40'
  # Using eval with arbitrary user input can be a security risk.
  # Ensure 'values' are numbers and 'operator' is a trusted arithmetic symbol.
  expression_parts = [str(val) for val in values]
  expression = operator.join(expression_parts)

  try:
    result = eval(expression)
    print(result)
    return result
  except Exception as e:
    print(f"Error evaluating expression '{expression}': {e}")
    return None
```

```python
from functools import reduce

# Lambda functions for arithmetic operations
cadd = lambda *args: sum(args)
# reduce with an initial value or check for empty args
csub = lambda *args: reduce(lambda x, y: x - y, args) if args else 0
cmul = lambda *args: reduce(lambda x, y: x * y, args) if args else 1
cdiv = lambda *args: (reduce(lambda x, y: x / y, args) if args and all(y != 0 for y in args[1:]) else "Error: Division by zero or insufficient arguments")

def calculator_lambda(*nums):
  if not nums: # Check if nums is empty
    print("Error: No arguments provided to calculator_lambda.")
    return None

  operator = nums[-1]
  values = nums[:-1]

  # Validate that all values are numeric
  for val in values:
    if not isinstance(val, (int, float)):
      print(f"Error: Non-numeric value '{val}' provided for calculation.")
      return None

  if not values and operator not in ['+', '*']:
    print(f"Error: No numbers provided for operation '{operator}'.")
    return None

  result = None # Initialize result

  if operator == '+':
    result = cadd(*values)
  elif operator == '-':
    if not values:
      print("Error: Subtraction requires at least one number.")
      return None
    result = csub(*values)
  elif operator == '*':
    result = cmul(*values)
  elif operator == '/':
    if not values:
      print("Error: Division requires at least one number.")
      return None
    if len(values) > 1 and any(v == 0 for v in values[1:]):
      print("Error: Division by zero is not allowed.")
      return None
    result = cdiv(*values)
  else:
    print(f"Error: Unsupported operator '{operator}'")
    return None

  if result is not None:
    print(f"Result: {result}")
  return result

print("\n--- Lambda Calculator Examples ---")
calculator_lambda(10, 20, 30, 40, '+')
calculator_lambda(10, 20, 30, 40, '-')
calculator_lambda(10, 20, 30, 40, '*')
calculator_lambda(10, 20, 30, 40, '/')
calculator_lambda(10, 0, '/') # Division by zero example
calculator_lambda() # No arguments example
calculator_lambda('a', 'b', '+') # Type error example
```

## Problem 3: String Clean up

Goal: Take a sentence, extract all words that do not contain vowels, and convert them to uppercase.

```python
vowels = {'a', 'e', 'i', 'o', 'u'}

# Split the text into words
words = text.split()

# Filter words that do not contain any vowels and convert them to uppercase
filtered_words_uppercase = [word.upper() for word in words if not any(vowel in word.lower() for vowel in vowels)]

print(filtered_words_uppercase)
```
