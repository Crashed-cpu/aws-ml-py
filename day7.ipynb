```python
class Student:
  def __init__(self, name, roll_no):
    self.name = name
    self.roll_no = roll_no

  def show_details(self):
    print(f'Name: {self.name}')
    print(f'Roll No: {self.roll_no}')

s = Student('him', 413)
s.show_details()
s.name
```

```python
class Student:
  def __init__ (this, name, roll_no, age):
    this.name = name
    this.roll_no = roll_no
    this.age = age

s = Student('nik', 232, 232)
print(s.name, s.roll_no, s.age)
```

```python
class Stu:
  def __init__(self,name,roll_no):
    self.name = name
    self.roll_no = roll_no
  # get
  def getName(self):
    print(self.name)
  # update
  def updateName(self, name=None):
    if name is None:
      print(self.name)
      return
    self.name = name
    print(self.name)
  # delete
  def deleteName(self):
    del self.name
    # print(self.name)
  # create
  def createName(self, fname):
    self.full_name = fname
    print(self.full_name)

s = Stu("nikhil",123)
s.createName('ayush saini')
s.updateName()
s.updateName('hi')
```

```python
class Student:
  def __init__(self, name, roll_no):
    self.name = name
    self.roll_no = roll_no

  def getData(self):
    return self.name, self.roll_no

# s = Student('ayu', 25)
# print(*s.getData(), end='\n')

class Address(Student):
  def __init__(self, name, roll_no, city, state):
    self.city = city
    self.state = state
    super().__init__(name, roll_no)

  def getAddress(self):
    return self.city, self.state

a = Address('ayush', 25, 'jaipur', 'rajasthan')
print(*a.getAddress(), *a.getData(),sep='\n')
```
